# Hospital Patient Data Overview

## Objectives
- Analyze patient demographics.
- Identify common diseases.
- Examine treatment outcomes.
- Analyze hospital bills.
- Study patient length of stay.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
df = pd.read_csv("../data/cleaned_data.csv")
```

```
df.head()
```

```
df.info()
```

```
df.describe()
```

```
df.isnull().sum()
```

```
print("Rows and Columns:", df.shape)
```

```
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.show()
```

```
gender = df["Gender"].value_counts()
plt.figure(figsize=(6,4))
plt.bar(gender.index, gender.values)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Patients")
plt.show()
```

```
disease = df["Disease"].value_counts()
plt.figure(figsize=(8,5))
plt.bar(disease.index, disease.values)
plt.xticks(rotation=45)
plt.title("Disease Distribution")
plt.xlabel("Disease")
plt.ylabel("Patients")
plt.show()
```

```
outcome = df["Outcome"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(outcome.values, labels=outcome.index, autopct="%1.1f%%")
plt.title("Treatment Outcomes")
plt.show()
```

```
plt.figure(figsize=(8,5))
plt.hist(df["Hospital_Bill"], bins=10)
plt.title("Hospital Bill Distribution")
plt.xlabel("Hospital Bill")
plt.ylabel("Patients")
plt.show()
```

```
plt.figure(figsize=(8,5))
plt.hist(df["Length_of_Stay"], bins=10)
plt.title("Length of Stay")
plt.xlabel("Days")
plt.ylabel("Patients")
plt.show()
```

```
print("Average Age:", df["Age"].mean())
print("Average Hospital Bill:", df["Hospital_Bill"].mean())
print("Maximum Bill:", df["Hospital_Bill"].max())
print("Minimum Bill:", df["Hospital_Bill"].min())
print("Average Length of Stay:", df["Length_of_Stay"].mean())
```
